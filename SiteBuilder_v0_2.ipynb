```python
#Initialize the OpenAI model and the prompt template
#Test Method to check mlflow. Ignore it.
llm = OpenAI(temperature=0.9)
prompt = PromptTemplate(
    input_variables=["GenAI"],
    template="What is a good name for a company that makes {product}?",
)

# Create the LLMChain with the specified model and prompt
chain = LLMChain(llm=llm, prompt=prompt)

# Log the LangChain LLMChain in an MLflow run
with mlflow.start_run():
    logged_model = mlflow.langchain.log_model(chain, "langchain_model")

# Load the logged model using MLflow's Python function flavor
loaded_model = mlflow.pyfunc.load_model(logged_model.model_uri)

# Predict using the loaded model
print(loaded_model.predict([{"product": "colorful socks"}]))
```

```python
import re

tech_string  = output_skills
# Use regular expression to extract text without numbers
extracted_text = re.sub(r'\d+\.\s', '', output_skills)
# Split the extracted text into a list using line breaks
python_list = extracted_text.split("\n")
# Remove any leading or trailing whitespace from each item in the list
python_list = [item.strip() for item in python_list]
# Now, python_list contains the items without numbers as a Python list
print(python_list)

span_list = []

for item in python_list:
    if item.strip():  # Check if the item is not empty (after stripping whitespace)
        span_element = f"<span>{item}</span>"
        span_list.append(span_element)


# Now, span_list contains the non-empty items as HTML span elements
for span in span_list:
    print(span)


# Wrap all the span elements in a div tag
div_skills= "\n".join(span_list)
div_skills = f"<div id='skills'>{div_skills}</div>"
```
